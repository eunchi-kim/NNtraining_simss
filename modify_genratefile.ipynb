```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
genpro_path = os.path.join('SIMsalabim','Data')
genpro_files = list(glob.glob(os.path.join(genpro_path, '*d250nm*')))
header = "x\tG_ehp"
for genpro_file in genpro_files:
    data = np.loadtxt(genpro_file, delimiter=' ', skiprows=2)
    x = data[:, 0]
    G = data[:, -1]
    base_name = os.path.basename(genpro_file)
    output_filename = os.path.join(genpro_path, f"SIM_{base_name}")
    combined_data = np.column_stack((x, G))
    np.savetxt(output_filename, combined_data, header=header, comments='', delimiter='\t')
    
```
